```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("Crop Production With Rainfall.csv")
data = df.dropna()
data
```

```python
s=list(data["State_Name"].unique())
y=list(data["Crop_Year"].unique())
```

```python
rain=pd.DataFrame()
for st in s:
    for ye in y:
        data1=data[data["State_Name"]==st][data["Crop_Year"]==ye][:1]
        rain=pd.concat([rain,data1])
rain
```

```python
rain=rain.drop(["District_Name","Season","Crop","Area","Production"],axis=1)
rain=rain.reset_index(drop=True)
rain
```

```python
rain_dum = pd.get_dummies(rain)
x = rain_dum.drop("Rainfall",axis=1)
y = rain_dum[["Rainfall"]]
```

```python
x
```

```python
y
```

```python
model = RandomForestRegressor()
model.fit(x,y.values.ravel())
```

```python
outrain=pd.DataFrame()
for st in s:
    testrain=rain[rain["State_Name"]==st][:1]
    outrain=pd.concat([outrain,testrain])
outrain
```

```python
outrain["Crop_Year"]=2021
rainpre= outrain.drop("Rainfall",axis=1)
test = pd.get_dummies(rainpre)
```

```python
test
```

```python
predict=model.predict(test)
predict
```

```python
rainpre["Rainfall"]=predict
rainpre=rainpre.reset_index(drop=True)
```

```python
rainpre
```

```python
rainpre.to_csv('Rainfall Prediction.csv')
```
